```python
#creating a dataframe
import pandas as pd

data={
        "name":["swaroop","tiru",'srinadh','ashok','yogi'],
        "age":[20,19,22,21,20],
        'marks':[80,75,70,75,85]
    }
df=pd.DataFrame(data)
print(df)

```

```python
import matplotlib.pyplot as plt
plt.bar(df["name"],df["marks"],color="red")
plt.title("marks of students")
plt.xlabel("Student")
plt.ylabel("marks")
plt.show()
```

```python
plt.scatter(df['age'],df['marks'],color='blue',marker=".")
plt.show()
```

```python
plt.plot(df["age"],df['marks'],marker=".",linestyle="--")
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))
plt.plot(df["age"],df['marks'],marker=".",linestyle="--")
plt.scatter(df['age'],df['marks'],color='blue',marker=".")
plt.show()
```
